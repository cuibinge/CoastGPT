```python
import numpy as np

# 1. 加载.npy文件（allow_pickle=True 允许加载字典格式数据）
history = np.load(
    "./training_history.npy",
    allow_pickle=True
).item()  # .item() 将numpy数组转换为Python字典

# 2. 查看数据结构（键：指标名称，值：对应训练过程的数组）
print("文件中包含的指标：", list(history.keys()))
print("\n" + "="*50)

# 3. 查看具体指标的内容（示例：查看验证集经纬度MAE）
val_lonlat_mae = history["val_lonlat_mae"]
print("验证集经纬度MAE（前5个epoch）：", val_lonlat_mae)
print("验证集经纬度MAE（最后5个epoch）：", val_lonlat_mae)
print(f"验证集经纬度MAE平均值：{np.mean(val_lonlat_mae):.4f}")
print(f"验证集经纬度MAE最小值：{np.min(val_lonlat_mae):.4f}（第{np.argmin(val_lonlat_mae)+1}个epoch）")

# 4. 查看其他指标（可选，按需修改）
print("\n" + "="*50)
print("训练总损失：", history["train_total_loss"])
```
